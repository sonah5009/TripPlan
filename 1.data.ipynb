```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
df = pd.read_csv('./data/KC_495_LLR_ATRCTN_2023.csv')
print(df.columns)
```

```python
# save
selected_columns = ['POI_ID', 'POI_NM', 'CTPRVN_NM', 'SIGNGU_NM','LEGALDONG_NM', 'LC_LO', 'LC_LA']
df_selected_origin = df[selected_columns]
print(pd.isna(df_selected_origin).sum(), '\n')
df_selected_origin = df_selected_origin.dropna()
print(pd.isna(df_selected_origin).sum(), '\n')

df_selected_origin = df_selected_origin.drop_duplicates(subset='POI_NM')
df_selected_origin.to_csv("./data/unique_set.csv", sep=',', index=False)
```

```python
columns_to_drop = ['ID', 'LCLAS_NM', 'MLSFC_NM', 'BHF_NM', 'ASSTN_NM', 'GID_CD', 'LAST_CHG_DE', 'ORIGIN_NM', 'FILE_NM','LI_NM', 'RDNMADR_CD', 'RDNMADR_NM', 'BULD_NO']


df.drop(columns=columns_to_drop, inplace=True)
df.head(3)
```

```python

desired_values = ['먹거리/패션거리' , '유명관광지' ,'해수욕장']
df = df[df['CL_NM'].isin(desired_values)]
# df = df[df['CL_NM'] == 'N']

# columns_to_drop = ['CL_CD', 'CL_NM']
# df.drop(columns=columns_to_drop, inplace=True)
desired_values = ['서울특별시', '경기도', '부산광역시']
df = df[df['CTPRVN_NM'].isin(desired_values)]
# df_test = df[~df['CTPRVN_NM'].isin(desired_values)]

# selected_columns = ['POI_ID', 'POI_NM', 'CTPRVN_NM', 'SIGNGU_NM','LC_LO', 'LC_LA']
selected_columns = ['POI_ID', 'POI_NM', 'CTPRVN_NM', 'SIGNGU_NM','LEGALDONG_NM', 'LC_LO', 'LC_LA']

df_selected = df[selected_columns]

df_unique = df_selected.drop_duplicates(subset='POI_NM')
```
